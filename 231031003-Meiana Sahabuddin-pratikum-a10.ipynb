```python
# Fungsi Rekursif Factorial dengan perulangan
def faktorial(nilai):
  if nilai<=1:
    return 1
  else:
    return nilai*faktorial(nilai-1)

# Program utama
for i in range(11):
  print('%2d ! = %d' % (i,faktorial(i)))
```

Tugas (Buatlah program dengan output sebagai berikut(ket: gunakan input sebagai masukan, 11 merupakan input))

```python
def faktorial_recursive(n):
  if n == 0 or n == 11:
    return 1
  else:
    return n * factorial_recursive(n - 1)

def factorial_loop(n):
  result = 1
  for i in range(1, n + 1):
    result *= i
  return result

#Input
number = int(input("Masukkan Angka untuk menghitung faktorial: "))

#Output
print(f"Hasil faktorial menggunakan rekursi: {faktorial_recursive(number)}")
print(f"Hasil faktorial menggunakan perulangan: {factorial_loop(number)}")
```
